```python
import attrs
import numpy as np
import xarray.tutorial

from eradiate.two.attrs import attrs_to_html_with_styles


@attrs.define
class MyClass:
    i = attrs.field()
    f = attrs.field()
    s = attrs.field()
    array = attrs.field()
    mapping = attrs.field()
    obj = attrs.field()
    ds = attrs.field()

    def _repr_html_(self):
        return attrs_to_html_with_styles(self)


MyClass(
    i=12345,
    f=12.345,
    s="hello world",
    array=np.zeros((5, 5, 5)),
    mapping={
        "foo": 0,
        "bar": 1,
        "qux": xarray.tutorial.load_dataset("air_temperature"),
    },
    obj=so,
    ds=xarray.tutorial.load_dataset("air_temperature"),
)
```
