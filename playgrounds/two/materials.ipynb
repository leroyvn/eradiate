```python
import drjit as dr
import mitsuba as mi
import numpy as np

import eradiate
from eradiate import KernelContext
from eradiate.two import Scene
from eradiate.two.materials import DiffuseMaterial
from eradiate.two.spectra import InterpolatedSpectrum
from eradiate.units import unit_registry as ureg

eradiate.set_mode("mono")
```

```python
s = InterpolatedSpectrum(
    wavelengths=np.linspace(300, 400, 11),
    values=np.linspace(0, 1, 11),
)
s
```

```python
mat = DiffuseMaterial(reflectance=s)
mat
```

```python
for w in [0.3, 0.35, 0.4] * ureg.um:
    ctx = KernelContext({"w": w})
    mat.update(ctx)

    si = dr.zeros(mi.SurfaceInteraction3f)
    print(mat.reflectance().eval(si))
```

```python
scene = Scene(bsdfs={"lambertian": mat})
scene.init()
print(scene.mi_scene)
scene
```
